```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# bir aylar listesi oluşturdum
aylar = ['Ocak', 'Şubat', 'Mart', 'Nisan',
       'Mayıs', 'Haziran', 'Temmuz', 'Ağustos', 'Eylül', 'Ekim', 'Kasım',
       'Aralık']

#veri setimi analizde daha rahat kullanabilmek için long formata çeviricem (melt işlemi ile)

df_long = df.melt(id_vars=["Kategori","Alt Kategori","Yıl"],
                  value_vars= aylar,
                  var_name= "Ay", value_name="Endeks")

df_long.head()
```

```python
df_long.shape
```

```python
#ay isimlerini numaraya çeviriyorum, mapping , ve ayrıca bir tarih sütunu oluşturuyorum
ay_map = {'Ocak': 1, 'Şubat': 2, 'Mart': 3, 'Nisan': 4,
       'Mayıs': 5, 'Haziran': 6, 'Temmuz': 7, 'Ağustos': 8, 
       'Eylül': 9, 'Ekim': 10, 'Kasım': 11,'Aralık': 12   
}

df_long["Ay_Num"] = df_long["Ay"].map(ay_map)
df_long["Tarih"] = pd.to_datetime(df_long["Yıl"].astype(str)+ "-" + df_long["Ay_Num"].astype(str))

df_long = df_long.sort_values(by="Tarih") #Yeni oluşturduğumuz Tarih sütununa göre veriyi kronolojik sıraya göre sıralıyor.
df_long.head()
```

```python
df_long = df_long.drop(columns=["Kategori"])   #çoğu analiz  işleminde kullanmaya gerek duymayacağımız  için baştan bu sütunu sildim
```

```python
#şimdi bu oluşturduğum long formatı csv olarak kaydediyorum
df_long.to_csv("tuik_tufe_long.csv", index=False, encoding="utf-8-sig")     #index=False: Veri çerçevesindeki index sütunu dosyaya yazılmaz.
                                                                            #encoding="utf-8-sig": Excel'de Türkçe karakter sorunu yaşamamak için önerilir.
```
